# Developing a Cryptography app in Python getting content from user 

- Program to **encrypt** a message and then **decrypt** it when the desired key is entered and also taking content from user using **getContentFromUser() function**.

```python
from cryptography.fernet import Fernet
```

```python
def generatePasskey():
    key = Fernet.generate_key()
    print(key)
    print(type(key))
    file = open("securityKey.key",'wb')
    file.write(key)
    file.close
```

```python
def getmykey():
    file = open("securityKey.key",'rb')
    return file.read()
```

```python
def getContentFromUser():
    return input("Enter the Content you want to Encrypt in your Python-Script")
```

```python
def encryptMessage(message_normal):
    key = getmykey()
    k = Fernet(key)   #fernet is an object in cryptography library used for encryption and decryption
    encrypted_Message = k.encrypt(message_normal)
    return encrypted_Message
```

```python
def decryptMessage(message_secret):
    key = getmykey()
    k = Fernet(key)
    decrypted_Message = k.decrypt(message_secret)
    return decrypted_Message
```

```python
generatePasskey()
```

```python
data = getContentFromUser()

encryptedcode = encryptMessage(data.encode('utf-8'))
```
